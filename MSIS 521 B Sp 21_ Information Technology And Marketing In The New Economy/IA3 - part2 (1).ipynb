```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import * # jaccard
from sklearn.metrics.pairwise import * # cosine
import pandas as pd
import nltk
from nltk.corpus import stopwords
import string
```

```python
#1
df = pd.read_excel('IA3-2.xlsx')
review = df.iloc[:,1]

lemmatizer = nltk.stem.WordNetLemmatizer()
stemmer = nltk.stem.PorterStemmer()

# lower_review = []
# for text in review:
#     tokenize = nltk.word_tokenize(text.lower())
#     lower_review.append(tokenize)
```

```python
processed_reviews = []
for text in review:
    white_token = nltk.tokenize.WhitespaceTokenizer()
    tokens = white_token.tokenize(text.lower())    
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if not token in stopwords.words('english') if token.isalpha()]
    tokens = [stemmer.stem(token) for token in tokens]
    processed_reviews.append(tokens)
processed_reviews
```

```python
strings=[]
for line in processed_reviews:
    string = ' '.join(line)
    strings.append(string)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2),min_df=5)
vectorizer.fit(strings)
review_vector = vectorizer.transform(strings)
```

```python
print(vectorizer.vocabulary_)
```

```python
#2
vectorizer2 = CountVectorizer(stop_words='english')
X2 = vectorizer2.fit_transform(strings)

from sklearn.decomposition import LatentDirichletAllocation
lda = LatentDirichletAllocation(n_components=6).fit(X2)
terms2 = vectorizer2.get_feature_names()

for idx, topic in enumerate(lda.components_):
    print('Topic %d:' % idx)
    print(" ".join([terms2[i] for i in topic.argsort()[:-4-1:-1]]))
    
lda.transform(X2[0])
```

```python
#3
import gensim

res_review = processed_reviews[1:10]

dictionary_r = gensim.corpora.Dictionary(res_review) 

bow_r = [dictionary_r.doc2bow(doc) for doc in res_review] 
lda_model_r = gensim.models.LdaModel(bow_r, num_topics=2, id2word=dictionary_r)
for idx, topic in lda_model_r.print_topics(-1, 4): 
    print('Restaurant Topic: {} \n Words: {}'.format(idx, topic))
    
print(lda_model_r[bow_r[2]])
```
